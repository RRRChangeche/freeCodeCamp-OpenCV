# Creating Panoramas using OpenCV

## Step for creating panoramas
1. Find keypoints in all images 
2. Find pairwise correspondences
3. Estimate pairwise homographies
4. Refine homographies
5. Switch with blending 

```python
import cv2
import glob
from matplotlib import pyplot as plt
import math
```

```python
imgfiles = glob.glob("./data/panoramas/*")
imgfiles.sort()

imgs = []
for img in imgfiles:
    img = cv2.imread(img)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    imgs.append(img)
```

```python
# Display imgs
num = len(imgs)
plt.figure(figsize=[30, 10])
cols = 3
rows = math.ceil(num/cols)

for i, img in enumerate(imgs):
    plt.subplot(rows, cols, i+1)
    plt.imshow(img)
```

```python
# Stitch images
stitcher = cv2.Stitcher_create()
status, result = stitcher.stitch(imgs)
if status == 0:
    plt.figure(figsize=[30,10])
    plt.imshow(result)
```
